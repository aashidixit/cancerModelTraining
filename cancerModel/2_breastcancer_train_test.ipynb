```python
bdatatraining = bdata.iloc[0:200,:]
mdatatraining=mdata.iloc[0:200,:]
```

```python
pcapb=bdatatraining.shape[0]/400
pcabm=mdatatraining.shape[0]/400
pcapb
```

```python
mucapb=bdatatraining['radius_mean'].mean()
mucapm=mdatatraining['radius_mean'].mean()
```

```python
sigmacapb=bdatatraining['radius_mean'].std()
sigmacapm=mdatatraining['radius_mean'].std()
```

```python
# covert the data into training set

# now compute this

# P(tumouris Benign/Stest)=
# (P(stest/tumour is Benign )*p(TUMOUR IS BEINGN))/
# (NUMERATOR + P(Stest/tumour is Malignant) * p(tumour is manignant)

# p(stest/tumour is benign)=
# (1/(sqrt(2*pi)*sigmabcap)) *exp(-(stestmucapb)**2/(2*sigmabcap**2))
# p(stest/tumour is malignant)=
# (1/(sqrt(2*pi)*sigmaMcap)) *exp(-(stestmucapM)**2/(2*sigmaMcap**2))
```

```python
bdatatesting=bdata.iloc[200:,:]
mdatatesting=mdata.iloc[200:,:]
```

```python
def UnivariateNaivebayesclassifier(oneexample):
    pstesttumourisbenign=s.norm.pdf(mucapb,sigmacapb,oneexample)
    numerator=pstesttumourisbenign * pcapb
    pstesttumourismanignant=s.norm.pdf(mucapm,sigmacapm,oneexample)
    secondpartofdenominator  =  pstesttumourismanignant * pcabm
    
    ptumourisbenignstest = (numerator)/(numerator + secondpartofdenominator) #posterior probability
    
    return ptumourisbenignstest
```

```python
truepositivecount = 0
falsepositivecount =0
truenegativecount = 0
falsenegativecount =0

for i in range(0,len(bdatatesting)):
    
    ptumourisbenignstest = UnivariateNaivebayesclassifier(bdatatesting.iloc[i,1])
    if ptumourisbenignstest > 0.5:
        truenegativecount +=1
    else:
        falsepositivecount +=1
        
```
